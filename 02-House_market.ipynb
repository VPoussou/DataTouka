4. Make an inner join of these four datasets into a single dataframe.

```python
main_df = house_market_df.merge(nb_rooms_df, on='id')
main_df = main_df.merge(nb_sdb_df, on='id')
main_df = main_df.merge(surface_area_df, on='id')

main_df.head()
```

5. What is the average surface area of the houses in our dataset?

```python
surface_mean = surface_area_df['superficie'].mean()
print(surface_mean)
```

6. What is the median and average number of rooms?

```python
mean_nb_rooms = nb_rooms_df['nombre_de_chambres'].mean()
median_nb_rooms = nb_rooms_df['nombre_de_chambres'].median()
print(mean_nb_rooms)
print(median_nb_rooms)
```

7. What is the average cost of a house?

```python
mean_house_price = house_market_df['house_price'].mean()
print(mean_house_price)
```

8. What is the average cost of a house, depending on the number of rooms it has?

```python
house_market_and_nb_rooms_df = house_market_df.join(nb_rooms_df, lsuffix='+')
mean_house_price_by_room_nb = house_market_and_nb_rooms_df.groupby('nombre_de_chambres').mean()
print(mean_house_price_by_room_nb)
```

Create five categories that respectively correspond to:
      - "very large" $=$ "a house larger than 25,000 sqrt_feet"
      - "large" $=$ "a house between 20,000 and 25,000 sqrt_feet"
      - "medium" $=$ "a house between 15,000 and 20,000 sqrt_feet"
      - "small" $=$ "a house between 10,000 and 15,000 sqrt_feet"
      - "very small" $=$ a house less than 10,000 sqrt_feet"

Fill in the corresponding rows in the house_size column.

```python
def thickk_evaluator(thickkness: float):
    if thickkness < 10_000:
        return 'very small'
    elif thickkness <= 15_000:
        return 'small'
    elif thickkness <= 20_000:
        return 'medium'
    elif thickkness <= 25_000:
        return 'large'
    elif thickkness > 25_000:
        return 'very large'
    
main_df['home_size'] = main_df['superficie'].apply(thickk_evaluator)

main_df.head()
```
